#Type

```python
x = 10
type(x)
print('x',type(x))

nome = 'Rodrigo'
print('nome',type(nome))

num = '100'
num_convert = int(num) #Converter variavel
print('num',type(num))
print('num_convert',type(num_convert))

num_float = float(num)
print('num_float',type(num_float))
#
print(num)
print(num_convert)
print(num_float)
#
```
