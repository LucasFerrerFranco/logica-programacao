#Condicional

```python
#If
if 5 < 10:
  print("Verdade")
```

```python
#if/else
num1 = 10
num2 = 5
if num1 > num2:
  print("O num1 é o maior número")
else:
  print("O num2 é o maior número")
```

```python
#if / elif/ else
num1 = 11
num2 = 10
if num1 > num2:
  print("O num1 é o maior número")
elif num2 > num1:
  print("O num2 é o maior número")
else:
  print("Os números são iguais")
```

#Condicional com operadores lógicos

```python
num1 = 9
num2 = 10
num3 = 2
if num1 > num2 and (num1 >num3):
  print("O num1 é o maior número")
else:
  print("O num1 não é o maior número")
```
